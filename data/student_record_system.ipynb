```python
import os
import csv
DATA_DIR="data"
os.makedirs(DATA_DIR,exist_ok=True)
csv_file=f"{DATA_DIR}/students.csv"
json_file=f"{DATA_DIR}/students.json"
def csv_write(students):
    with open(csv_file,'w',newline='') as file:
        writer=csv.DictWriter(file,fieldnames=['name','grade'])
        writer.writeheader()
        writer.writerows(students)
        

def csv_read():
    if not os.path.exists(csv_file):
        return []
    with open(csv_file,'r') as file:
        reader=csv.DictReader(file)
        return list(reader)
```

```python
students=csv_read()
while True:
    print("1.add student")
    print("2.view all students")
    print("3.save & stop")
    ch=int(input("enter option : "))
    if ch==1:
        stu={
        'name':input('enter name'),
        'grade':input('enter grade')
        }
        students.append(stu)
    elif ch==2:
        for s in students:
            print(s)
    elif ch==3:
        csv_write(students)
        break
    else:
        print('invalid')
```
